# Gantt chart
### Author: Szymon Nowakowski



```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define project timeline
start_date = datetime(2026, 7, 1)
end_date = datetime(2030, 6, 30)

# Define tasks and their durations
tasks = [
    ("Task 1", start_date, start_date + timedelta(days=4*365)),
    ("Task 2", start_date + timedelta(days=365 * 1), start_date + timedelta(days=366 * 1.5)),
    ("Task 3", start_date + timedelta(days=366 * 1.5), start_date + timedelta(days=366 * 2.5)),
    ("Task 4", start_date + timedelta(days=366 * 2.5), start_date + timedelta(days=365 * 4)),
    ("Task 5", start_date + timedelta(days=365 * 1.5), start_date + timedelta(days=366 * 2.5)),
]

fig, ax = plt.subplots(figsize=(10, 6))
y_pos = range(len(tasks))

for i, (task, start, end) in enumerate(tasks):
    start_num = mdates.date2num(start)
    end_num = mdates.date2num(end)
    ax.barh(i, end_num - start_num, left=start_num)
    ax.text(start_num + (end_num - start_num) / 2, i, task,
            ha='center', va='center')

# Formatting
ax.invert_yaxis()
ax.set_yticks(list(y_pos))
ax.set_yticklabels([task[0] for task in tasks])

ax.set_xlim(mdates.date2num(start_date), mdates.date2num(end_date))
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

ax.set_xlabel("Year")
ax.set_ylabel("Tasks")
ax.set_title("Gantt Chart for Project Tasks")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
